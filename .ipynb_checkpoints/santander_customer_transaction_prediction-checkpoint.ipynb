## 3. Bayesian Linear Regression 
<div class="alert alert-info">
a) Implement Bayesian linear regression<br>
    
</div>

<div class="alert alert-success">
The Bayesian Linear Regression is based on the Bayes rule: $$ p(w \mid X,t) = \frac{p(t\mid X,w) p(w)}{p(t\mid X)} $$ We know that $p(t \mid w, X, \sigma^2) = N(Xw, \sigma^2I)$, the prior conjugate to gaussian is gaussian so we choose $p(w) = N(0,S)$ (where $0$ and $S$ are design choices). So posterior must be gaussian $N(\mu, \Sigma)$, with: $$ \Sigma = (\frac{1}{\sigma^2}X^TX + S^{-1})^{-1}$$ and $$ \mu = \frac{1}{\sigma^2}\Sigma X^Tt $$
</div>

```python
class LinearBayes(object):
    """
    A class that holds parameter prior/posterior and handles 
    the hyper-parameter updates with new data
        
    """
    def __init__(self, _mean, _S, sigma):
        # Hyper parameter sigma
        self.noise_sigma = 1/sigma**2
        # Vector of matrix mean
        self.mean = _mean.reshape(_mean.shape + (1,))
        # Covariance S
        self.S = _S 
    
    def get_design_matrix(self, x):
        X = np.ones(x.shape[0], x.shape[1] + 1)
        X[:, 1:] = x
        X
        return X

    def fit(self, X, target):
        """
        Updates given vectors of x-values and t-values
        """
        # Need to convert target from an array into a column vector
        # to correctly compute matrix multiplication
        t = target.reshape(target.shape + (1,))
        
        self.cov = np.linalg.inv(np.linalg.inv(self.S) + self.noise_sigma*X.T.dot(X)) #Covaricance sigma
        self.mean = self.cov.dot(self.noise_sigma*X.T.dot(t))
        
    def get_predictions(self, x_new):
        t_new = x_new.dot(self.mean)
        var = [self.noise_sigma +x_n.dot(self.S.dot(x_n.T)) for x_n in x_new]
        return t_new, var
```

c) (code) Write code that calculates the N-th order polynomial transformation of the
input data. For simplicity, do not consider polynomials of more than one variable
(such as x<sup>2</sup>y), but raise each input variable to the power of N individually.
Consider N=1, 2, 3, and 6. 
</div>

```python
def polynomial_transformation(x, order):
    # ALLOCATE MATRIX
    X = np.ones(shape=(x.shape[0], x.shape[1]), dtype=int)
    # STACK COLUMNS
    for i in range(order):
        X = np.hstack((X, np.power(x, i+1))) 

    return X
```

## 4. Logistic Regression
<div class="alert alert-info">
a) (code) The goal is to implement a logistic regression classifier that optimizes for the
Maximum a Posteriori (MAP) estimate; assume a Gaussian prior on the parameters.
As a first step, write a function that calculates the gradient of the joint likelihood. <br>
    
</div>

```python
class LogisticRegression(object):
    def __init__(self, _mean, _S):
        # Vector of matrix mean
        self.w = _mean.reshape(_mean.shape + (1,))
        # Covariance S
        self.S = _S 
        # learning rate
        self.lr
        # threhold
        self.threshold = 0.0001
        
    def sigmoid(z):
        return 1 / (1 + np.exp(-z))
    
    def get_cost( y, y_hat):
        return - np.sum(np.dot(y.T,np.log(1-y_hat)+ np.dot((1-y).T,np.log(1-y_hat)))) / ( len(y))

    def get_gradient(self, x, target):
        t = target.reshape(target + (1,))
        
        return x.T.dot(t - sigmoid(x.dot(self.w)))
    
    def joint_likelihood(self, x, target):
        t = target.reshape(target + (1,))
        lv = (1-t)*(np.exp(-x.dot(self.w))/(1+np.exp(-x.dot(self.w)))) + t/(1+np.exp(-x.dot(self.w)))
        return np.sum(np.log(lv))
    
    def likelihood_predictions(self, x, target):
        if (target == 1):
            t = sigmoid(x.dot(self.w))
        else:
            t = 1 - sigmoid(x.dot(self.w))
        return t.flatten()
    
    def fit(self, x, t):
        pre_likelihood = self.joint_likelihood(x, t)
        self.w = self.w + self.lr*self.get_gradient(x, t)
        joint_likelihood = self.joint_likelihood(x, t)
        i=1
        while(joint_likelihood - pre_likelihood > self.threshold and i <500):
            pre_likelihood = joint_likelihood
            self.w = self.w + self.lr*self.get_gradient(x, t)
            joint_likelihood = self.joint_likelihood(x, t)
            i = i + 1
    
    def get_predictions(self, x):
        t = 1/sigmoid(x.dot(self.w))
        t_new = t.flatten()
        t_new[t_new >= 0.5] = 1
        t_new[t_new < 0.5] = 0
        
        return t_new
        
        
```


<div class="alert alert-info">
c) (text) Comment on the convexity of the problem; do you need multiple restarts in order
to obtain a solution sufficiently close to the global optimum? <br>
    
</div>

We have $ p=P(y=1|x,w) = \frac{1}{1+exp(-wx)} $
and
$ P(y=0|x,w) = (1-p)= \frac{1}{1+exp(wx)} $
The posterior function is updated by Bayesian Logistic Regression parameters w. So we can maximize poterior:
$$argmax_k log\left( \sum_{i=1}^n y_ip_i+(1-y_i)(1-p_i) \right) - \frac{1}{2}(w-w_0)^T S_0^-1 (w-m_0)$$
The log prior function $\frac{1}{2}(w-w_0)^T S_0^-1 (w-m_0)$ is concave.<br>
The log likelihood is concave: (1)
$$\frac{\partial P(w|X,y)}{\partial w_j} = \sum_{i=1}^n x_{ij}(y_i-p_i) $$
$$\frac{\partial^2 P(w|X,y)}{\partial w_j} = -\sum_{i=1}^n x_{ij} x_{ik} p_i(1-p_i)$$
    The Hessian is negative semi-definite, the nxn matrix A ia said to be negative semidefinite if for all nx1 vextor y<>0: y^TAy. We can clarify the Hessian as: (2)
    $$y^TAy = - \sum_{i=1}^n y^Tp_ip_i^Ta = -(a^Tp_i)^2 <= 0 $$
 From (1)(2), we can see that the convexity of the problem is concave so there is only 1 local maximum and we do not need multiple restarts in order to obtain a solution sufficiently close to the global optimum.
